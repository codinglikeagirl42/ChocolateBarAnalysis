#### Tastes

I wanted to be able to take a subset of data originally it was just get rid of all the null values from the data set which leads me 242 records, but I reexamined the percentage totals of the null values and decided too make two new CSVs, to see how the information compared. One with deleting the fourth_taste column and then getting rid of any row with null values, leaves 1604 records.  The other keeping the fourth_taste and deleting any row with null values.

```python
# From https://towardsdatascience.com/data-cleaning-in-python-the-ultimate-guide-2020-c63b88bf0a0d
# How complete is my code
# % of missing.

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Third_Taste
# Create an new dataframe that is a subset of the chocolate_cleaned, delete the fourth taste row 
tastes3_df = df.drop(columns=['fourth_taste'], axis=1)

# then any rows with null values for tastes 1-3
tastes3_df = tastes3_df.dropna()

#save dataset to a new csv
tastes3_df.to_csv("tastes3.csv")

# Check to see that it workded
tastes3 = pd.read_csv("tastes3.csv")


# Want to see if the dataframe was properly imported and see what kind of data is stored in each column.
tastes3.info()
tastes3.head()
```

```python
# Create an new dataframe that is a subset of the chocolate_cleaned withoug any rows with null values for tastes

tastes4_df = df.dropna()

#save dataset to a new csv
tastes4_df.to_csv("tastes4.csv")

# Check to see that it workded
tastes4 = pd.read_csv("tastes4.csv")


# Want to see if the dataframe was properly imported and see what kind of data is stored in each column.
tastes4.info()
tastes4.head()
```

#### How does the average rating % change between the different datasets

```python
# Average Ratings
# Some variation across datasets, wonder if it's statistically significant
df_t = df['rating'].mean()
df_t3 = tastes3['rating'].mean()
df_t4 = tastes4['rating'].mean()

print('Full DF: ' + str(df_t))
print('3 Tastes No Null: ' + str(df_t3))
print('4 Tastes No Null: ' + str(df_t4))
```

#### How does the rating % change between the different datasets

```python
# Ratings
# Some variation across datasets, wonder if it's statistically significant
df_t = df['rating'].value_counts(normalize=True).sort_index(ascending=False).mul(100).round(1).astype(str) + '%'
df_t3 = tastes3['rating'].value_counts(normalize=True).sort_index(ascending=False).mul(100).round(1).astype(str) + '%'
df_t4 = tastes4['rating'].value_counts(normalize=True).sort_index(ascending=False).mul(100).round(1).astype(str) + '%'

pd.DataFrame({'Full DF': df_t, '3 Tastes No Null': df_t3, '4 Tastes No Null': df_t4})
```

```python
#https://www.delftstack.com/howto/numpy/python-numpy-value-counts/
array = np.array(tastes4[['first_taste', 'second_taste', 'third_taste', 'fourth_taste']])

unique, counts = np.unique(array, return_counts=True)

result = np.column_stack((unique, counts)) 
print (result)
```
